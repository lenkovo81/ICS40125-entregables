```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')

# Convertir 'date_added' a formato de fecha y extraer año y mes
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', format='mixed')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()

df.head()
```

### 4. Número de títulos agregados por año

```python
# Contar títulos agregados por año
titles_by_year = df['year_added'].value_counts().sort_index().dropna()

plt.figure(figsize=(12, 6))
sns.lineplot(x=titles_by_year.index, y=titles_by_year.values)
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Títulos Agregados')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

**Respuesta:** El catálogo de Netflix experimentó su mayor crecimiento entre los años **2016 y 2019**, con un pico en 2019. Esto coincide con el período de fuerte expansión internacional de la plataforma, cuando Netflix lanzó su servicio en más de 130 nuevos países a principios de 2016. A medida que se expandía globalmente, la plataforma necesitaba agregar una gran cantidad de contenido para satisfacer a las nuevas audiencias y establecer su presencia en diversos mercados.

### 5. Heatmap de lanzamientos por año y mes

```python
# Crear tabla pivote para el heatmap
heatmap_data = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)

# Reordenar los meses para una visualización correcta
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
               'August', 'September', 'October', 'November', 'December']
heatmap_data = heatmap_data[month_order]

plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5, linecolor='black')
plt.title('Títulos Agregados a Netflix por Año y Mes')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.show()
```

#### Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

**Respuesta:** Sí, se observa una clara estacionalidad en los lanzamientos. Los meses de **Julio, Octubre, Noviembre y Diciembre** tienden a tener un número significativamente mayor de títulos agregados, especialmente en los años de mayor crecimiento (2018-2019). Esto podría deberse a estrategias de lanzamiento vinculadas a períodos vacacionales, festivales de cine, o el fin de año, cuando la gente tiene más tiempo libre para consumir contenido. Por otro lado, los meses de **Febrero y Marzo** suelen mostrar una menor actividad.

### 6. Duración de películas por género

```python
# Filtrar solo películas
movies_df = df[df['type'] == 'Movie'].copy()

# Extraer la duración en minutos y convertir a numérico
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '', regex=False).astype(float)

# Para manejar múltiples géneros, dividimos la cadena y 'expandimos' el DataFrame
movies_df['listed_in'] = movies_df['listed_in'].apply(lambda x: x.split(', ')) # Convertir cadena a lista de géneros
m_genres = movies_df.explode('listed_in')
m_genres = m_genres.reset_index(drop=True) # Resetear el índice para evitar duplicados

# Agrupar por género y crear boxplot de duración
plt.figure(figsize=(14, 8))
sns.boxplot(data=m_genres, y='listed_in', x='duration_minutes', palette='tab10',
            order=m_genres.groupby('listed_in')['duration_minutes'].median().sort_values(ascending=False).index,
            hue='listed_in', legend=False) # Añadido hue y legend=False
plt.title('Distribución de Duración de Películas por Género')
plt.xlabel('Duración (minutos)')
plt.ylabel('Género')
plt.show()
```

#### Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?

**Respuesta:** Los géneros que suelen tener una mayor duración promedio de películas incluyen **Documentaries, Dramas, Independent Movies y Foreign Films**. Esto es coherente con el tipo de narrativas y profundidad que suelen abordar. Por otro lado, géneros como **Children & Family Movies** tienden a ser más cortos.

En cuanto a la variabilidad, muchos géneros muestran una **alta dispersión** en la duración (rangos intercuartílicos amplios y muchos *outliers*), lo que indica que, si bien hay una duración típica, también existen películas mucho más cortas o mucho más largas dentro del mismo género. Por ejemplo, los Dramas y Documentaries tienen una gran variabilidad, sugiriendo que hay tanto producciones cortas como largometrajes extensos. Géneros como Comedies también muestran una variabilidad considerable, aunque con una mediana más baja.

### 8. Películas vs Series según género

```python
# Para manejar múltiples géneros, dividimos la cadena y 'expandimos' el DataFrame
genres_df = df.copy()
genres_df['listed_in'] = genres_df['listed_in'].apply(lambda x: x.split(', ')) # Convertir cadena a lista de géneros
genres_exploded = genres_df.explode('listed_in')

# Contar títulos por tipo y género
genre_type_counts = genres_exploded.groupby(['listed_in', 'type']).size().unstack(fill_value=0)

# Seleccionar los géneros más relevantes para no saturar el gráfico
top_genres = genres_exploded['listed_in'].value_counts().head(15).index
genre_type_counts_top = genre_type_counts.loc[top_genres].sort_values(by='Movie', ascending=False)

plt.figure(figsize=(14, 8))
genre_type_counts_top.plot(kind='barh', stacked=True, figsize=(14, 8), cmap='Paired')
plt.title('Distribución de Películas y Series por Género')
plt.xlabel('Cantidad de Títulos')
plt.ylabel('Género')
plt.legend(title='Tipo')
plt.tight_layout()
plt.show()
```

#### Pregunta guía: ¿Existen géneros más asociados a películas o a series?

**Respuesta:** Sí, existen claras asociaciones de géneros con películas o series. Por ejemplo:

*   **Películas:** Géneros como 'Dramas', 'Comedies', 'Documentaries', 'International Movies', 'Independent Movies' y 'Action & Adventure' muestran una cantidad predominante de películas en comparación con series.
*   **Series:** Por otro lado, géneros como 'International TV Shows', 'Crime TV Shows', 'TV Dramas', 'Kids' TV', 'Romantic TV Shows', 'TV Comedies' y 'Docuseries' tienen una proporción significativa o mayor de series. Esto resalta que Netflix invierte en series de televisión en categorías específicas para atraer a audiencias que prefieren este formato, especialmente en el ámbito internacional y de nicho.

### 9. Relación entre duración y año de lanzamiento

```python
# Filtrar solo películas para este análisis (ya que solo ellas tienen duración en minutos)
movies_only_df = df[df['type'] == 'Movie'].copy()

# Convertir la duración a minutos, manejando posibles errores y valores nulos
movies_only_df['duration_minutes'] = movies_only_df['duration'].str.replace(' min', '', regex=False)
movies_only_df['duration_minutes'] = pd.to_numeric(movies_only_df['duration_minutes'], errors='coerce')
movies_only_df.dropna(subset=['duration_minutes'], inplace=True)

plt.figure(figsize=(15, 8))
sns.scatterplot(data=movies_only_df, x='release_year', y='duration_minutes', alpha=0.6, s=50)
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?

**Respuesta:** Observando el gráfico de dispersión, parece que la **duración promedio de las películas se ha mantenido relativamente estable** a lo largo de los años, rondando generalmente entre los 80 y 120 minutos. Sin embargo, se puede notar una mayor concentración de lanzamientos en años más recientes, lo que ya habíamos identificado.

Sí, se observan **outliers interesantes**: hay películas muy cortas (por debajo de los 30-40 minutos) y películas muy largas (por encima de los 180-200 minutos) distribuidas a lo largo de todo el rango de años, aunque con una mayor frecuencia en los años más recientes. Esto sugiere que, si bien la media se mantiene, hay una tendencia a experimentar con formatos más cortos (cortometrajes, especiales) y producciones épicas más largas.

### Desafío Final: Combinaciones más frecuentes de género + rating

```python
# Para manejar múltiples géneros, dividimos la cadena y 'expandimos' el DataFrame
genres_ratings_df = df.copy()
genres_ratings_df['listed_in'] = genres_ratings_df['listed_in'].apply(lambda x: x.split(', ')) # Convertir cadena a lista de géneros
genres_ratings_exploded = genres_ratings_df.explode('listed_in')

# Eliminar filas con valores nulos en 'rating' o 'listed_in' (si los hubiera después del explode)
genres_ratings_exploded.dropna(subset=['rating', 'listed_in'], inplace=True)

# Calcular la frecuencia de cada combinación de género y rating
combination_counts = genres_ratings_exploded.groupby(['listed_in', 'rating']).size().unstack(fill_value=0)

# Para un heatmap legible, seleccionaremos los ratings más comunes y los géneros más comunes
top_ratings = genres_ratings_exploded['rating'].value_counts().head(10).index
top_genres_for_heatmap = genres_ratings_exploded['listed_in'].value_counts().head(20).index

heatmap_data_filtered = combination_counts.loc[top_genres_for_heatmap, top_ratings]

plt.figure(figsize=(16, 10))
sns.heatmap(heatmap_data_filtered, cmap='Blues', annot=True, fmt='d', linewidths=.5, linecolor='black')
plt.title('Frecuencia de Combinaciones Género + Clasificación por Edad')
plt.xlabel('Clasificación por Edad')
plt.ylabel('Género')
plt.show()
```
