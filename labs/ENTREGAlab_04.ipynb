### Respuesta a la pregunta e):

### Países con índice constante

Sí, se identificaron **Antigua y Barbuda** y **Granada** como los países con un `indice` constante (sin variabilidad) a lo largo de los años registrados con datos.

### Países sin datos (todos los valores en cero)

No se encontraron países que no tuvieran ningún dato (es decir, con todos los valores en 0) en la tabla dinámica. Todos los países tienen al menos un dato registrado.

````python
import numpy as np

print("--- Tabla dinámica (pivot_df) de índice anual por país ---")
pivot_df = pd.pivot_table(df, values='indice', index='pais', columns='anio', aggfunc='max', fill_value=0)
display(pivot_df.head())

print("\n--- Análisis de Tabla Dinámica (Extremos) ---")
max_indice_global = pivot_df.max().max()
min_indice_global = pivot_df[pivot_df > 0].min().min() # Excluir ceros

max_indice_pais_anio = pivot_df.stack().idxmax()
max_pais = max_indice_pais_anio[0]
max_anio = max_indice_pais_anio[1]

min_indice_pais_anio = pivot_df[pivot_df > 0].stack().idxmin()
min_pais = min_indice_pais_anio[0]
min_anio = min_indice_pais_anio[1]

print(f"País con el mayor 'indice' en toda la tabla: {max_pais} en el año {max_anio} con un indice de {max_indice_global}")
print(f"País con el menor 'indice' (distinto de cero) en toda la tabla: {min_pais} en el año {min_anio} con un indice de {min_indice_global}")

print("\n--- Análisis de Tabla Dinámica (Años) ---")
promedio_indice_por_anio = pivot_df.mean(axis=0)

anio_mas_alto_promedio = promedio_indice_por_anio.idxmax()
valor_mas_alto_promedio = promedio_indice_por_anio.max()

anio_mas_bajo_promedio = promedio_indice_por_anio.idxmin()
valor_mas_bajo_promedio = promedio_indice_por_anio.min()

print(f"Año con el promedio de 'indice' más alto: {anio_mas_alto_promedio} (Promedio: {valor_mas_alto_promedio:.2f})")
print(f"Año con el promedio de 'indice' más bajo: {anio_mas_bajo_promedio} (Promedio: {valor_mas_bajo_promedio:.2f})")

print("\n--- Análisis de Tabla Dinámica (Variabilidad) ---")
max_indice_por_pais = pivot_df.replace(0, np.nan).max(axis=1)
min_indice_por_pais = pivot_df.replace(0, np.nan).min(axis=1)

variabilidad_por_pais = max_indice_por_pais - min_indice_por_pais

pais_mayor_variabilidad = variabilidad_por_pais.idxmax()
valor_mayor_variabilidad = variabilidad_por_pais.max()

print(f"País con mayor variabilidad en 'indice': {pais_mayor_variabilidad} (Variabilidad: {valor_mayor_variabilidad:.2f})")

print("\n--- Análisis de Tabla Dinámica (Constancia y Nulos) ---")
paises_indice_constante = variabilidad_por_pais[variabilidad_por_pais == 0].index.tolist()

print("Países con índice constante (sin variabilidad a lo largo de los años con datos):")
if paises_indice_constante:
    for pais in paises_indice_constante:
        print(f"- {pais}")
else:
    print("Ningún país tiene un índice constante.")

paises_sin_datos = pivot_df[pivot_df.sum(axis=1) == 0].index.tolist()

print("\nPaíses sin ningún dato (todos los valores en 0):")
if paises_sin_datos:
    for pais in paises_sin_datos:
        print(f"- {pais}")
else:
    print("Todos los países tienen al menos un dato.")
````

## Comparación Regional (Bucle For)

### Subtask:
Para cada año, en el subconjunto de países latinoamericanos, identificar el país con el menor y mayor valor de 'indice'.


**Reasoning**:
I need to filter the DataFrame `df` to include only the countries listed in the `america` variable. I will use the `isin()` method on the `codigo_iso` column for this purpose and store the result in `df_america`.



````python
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
           'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
           'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
           'USA', 'VEN']

df_america = df[df['codigo_iso'].isin(america)].copy()

print("DataFrame filtrado para países latinoamericanos (df_america) creado.")
print(f"Shape de df_america: {df_america.shape}")
df_america.head()
````

**Reasoning**:
I will iterate through each unique year in `df_america`, filter the data for that year, and then find the countries with the minimum and maximum 'indice' values.



````python
años_unicos = df_america['anio'].sort_values().unique()

print("Comparación Regional (por año) usando un bucle for:")
for anio in años_unicos:
    df_anio_america = df_america[df_america['anio'] == anio].dropna(subset=['indice'])

    if not df_anio_america.empty:
        # País con el menor índice (mayor libertad de prensa)
        min_indice_pais = df_anio_america.loc[df_anio_america['indice'].idxmin()]

        # País con el mayor índice (menor libertad de prensa)
        max_indice_pais = df_anio_america.loc[df_anio_america['indice'].idxmax()]

        print(f"\nAño {anio}:")
        print(f"  Mayor libertad de prensa: {min_indice_pais['pais']} (indice: {min_indice_pais['indice']})")
        print(f"  Menor libertad de prensa: {max_indice_pais['pais']} (indice: {max_indice_pais['indice']})")
    else:
        print(f"\nAño {anio}: No hay datos de 'indice' disponibles para este año en la región.")
````

## Comparación Regional (Vectorizado)

### Subtask:
Resolver la misma tarea de comparación regional para países latinoamericanos utilizando un enfoque vectorizado con `groupby`, sin usar ciclos explícitos.


**Reasoning**:
I will group `df_america` by 'anio' and then use `idxmin()` and `idxmax()` on the 'indice' column to find the rows corresponding to the countries with the minimum and maximum indices for each year. This will allow for a vectorized approach as requested.



````python
print("Comparación Regional (por año) usando un enfoque vectorizado:")

# Filtrar NaN en 'indice' antes de agrupar para evitar errores en idxmin/idxmax
df_america_filtered = df_america.dropna(subset=['indice'])

# Encontrar el país con el menor índice por año (mayor libertad de prensa)
idx_min = df_america_filtered.groupby('anio')['indice'].idxmin()
paises_mayor_libertad = df_america_filtered.loc[idx_min]

# Encontrar el país con el mayor índice por año (menor libertad de prensa)
idx_max = df_america_filtered.groupby('anio')['indice'].idxmax()
paises_menor_libertad = df_america_filtered.loc[idx_max]

# Mostrar los resultados
for anio in paises_mayor_libertad['anio'].unique():
    mayor_libertad_pais = paises_mayor_libertad[paises_mayor_libertad['anio'] == anio].iloc[0]
    menor_libertad_pais = paises_menor_libertad[paises_menor_libertad['anio'] == anio].iloc[0]

    print(f"\nAño {anio}:")
    print(f"  Mayor libertad de prensa: {mayor_libertad_pais['pais']} (indice: {mayor_libertad_pais['indice']})")
    print(f"  Menor libertad de prensa: {menor_libertad_pais['pais']} (indice: {menor_libertad_pais['indice']})")

# Handle years with no 'indice' data, similar to the loop version
años_con_datos = df_america_filtered['anio'].unique()
años_unicos_total = df_america['anio'].sort_values().unique()

for anio in años_unicos_total:
    if anio not in años_con_datos:
        print(f"\nAño {anio}: No hay datos de 'indice' disponibles para este año en la región.")
````
