## Revisar y Describir Dataset

### Subtarea:
Revisar y describir el dataset de Netflix, incluyendo el número de filas y columnas, los tipos de datos de cada columna y la cantidad de valores nulos por columna.

```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')

print(f"El DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

print("Tipos de datos de cada columna:\n")
df.info()

print("\nCantidad de valores nulos por columna:\n")
print(df.isnull().sum())
```

## Transformar y Crear Columnas de Fecha

### Subtarea:
Convertir la columna 'date_added' al tipo de dato fecha y luego crear dos nuevas columnas, 'year_added' y 'month_added', extrayendo el año y el mes de 'date_added' respectivamente.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

print("Columnas 'date_added', 'year_added' y 'month_added' actualizadas/creadas.")
print("Primeras filas con las nuevas columnas:")
print(df[['date_added', 'year_added', 'month_added']].head())
```

## Seleccionar Películas Post-2018

### Subtarea:
Utilizar `.loc` para seleccionar todas las entradas que son de tipo 'Movie' y que fueron añadidas después del año 2018.

```python
movies_post_2018 = df.loc[(df['type'] == 'Movie') & (df['year_added'] > 2018)]
print("Películas agregadas después de 2018:")
print(movies_post_2018.head())
```

## Filtrar por 'love' y Extraer Duración

### Subtarea:
Filtrar los títulos que contienen la palabra 'love' (sin importar mayúsculas/minúsculas) usando `str.contains()`. Luego, extraer la duración en minutos para las películas desde la columna 'duration' utilizando `str.extract()`.

```python
df_love_titles = df[df['title'].str.contains('love', case=False, na=False)].copy()

df_love_titles.loc[df_love_titles['type'] == 'Movie', 'duration_minutes'] = (
    df_love_titles[df_love_titles['type'] == 'Movie']['duration']
    .str.extract(r'(\d+)\smin')
    .astype(float)
)

print("Primeras filas de los títulos con 'love' y la duración en minutos:")
print(df_love_titles[['title', 'duration', 'duration_minutes']].head())
```

## Explode y Top 10 Géneros

### Subtarea:
Aplicar `explode()` sobre la columna `listed_in` para obtener una fila por cada género. Luego, obtener un top 10 de géneros más frecuentes utilizando `value_counts()`.

```python
df_exploded_genres = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
top_10_genres = df_exploded_genres['listed_in'].value_counts().head(10)

print("Top 10 Géneros más frecuentes:")
print(top_10_genres)
```

## Marcar Películas Largas

### Subtarea:
Crear una nueva columna para marcar las películas con una duración superior a 120 minutos como 'contenido largo' utilizando `where()` y `mask()`.

```python
df['duration_minutes'] = df['duration'].str.extract(r'(\d+)\smin').astype(float)

df['long_content'] = None
df['long_content'] = df['long_content'].mask(
    (df['type'] == 'Movie') & (df['duration_minutes'] > 120),
    'contenido largo'
)

print("Primeras filas con las nuevas columnas 'duration_minutes' y 'long_content':")
print(df[['title', 'type', 'duration', 'duration_minutes', 'long_content']].head(10))
```

## Filtrar Películas Específicas

### Subtarea:
Utilizar `.loc` para filtrar películas que cumplen con las siguientes condiciones: más de 100 minutos de duración, clasificación ('rating') igual a 'R', y país de origen igual a 'United States'.

```python
filtered_movies_specific = df.loc[
    (df['duration_minutes'] > 100) &
    (df['rating'] == 'R') &
    (df['country'] == 'United States')
]

print("Películas que cumplen con las condiciones especificadas:")
print(filtered_movies_specific.head())
```

## Formatear Top 10 Películas Más Largas

### Subtarea:
Utilizar `.style` para formatear visualmente el top 10 de películas más largas, destacando la duración.

```python
movies_df = df[df['type'] == 'Movie'].copy()

top_10_longest_movies = movies_df.sort_values(by='duration_minutes', ascending=False).head(10)

print("Top 10 Películas Más Largas (Formato Estilizado):")
top_10_longest_movies_styled = top_10_longest_movies.style.background_gradient(subset=['duration_minutes'], cmap='viridis')

top_10_longest_movies_styled
```

## Combinaciones Frecuentes de Género y Rating

### Subtarea:
Encontrar las combinaciones más frecuentes de género y rating en el dataset, utilizando `value_counts()` después de aplicar `explode()` a la columna de géneros.

```python
df_exploded_genres_with_rating = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

df_exploded_genres_with_rating['genre_rating_combination'] = \
    'Genre: ' + df_exploded_genres_with_rating['listed_in'] + \
    ', Rating: ' + df_exploded_genres_with_rating['rating'].fillna('Unknown')

top_10_genre_rating_combinations = df_exploded_genres_with_rating['genre_rating_combination'].value_counts().head(10)

print("Top 10 Combinaciones Más Frecuentes de Género y Rating:")
print(top_10_genre_rating_combinations)
```
